### Goal 
We will examine the effect of tree density, tree health, and biodiversity on the spread-rate of the Emerald Ash Borer. 

### Justification
Given the variable behavior and life stages of the Emerald Ash Borer and the importance of spatial considerations when modeling their behavior, an ABM methodology is necessary. A Differential Equations approach would not be prudent in this case, given the relatively stochastic ways in which the Ash Borer is introduced and subsequently spreads.

### Processes & Mechanisms of Interest
In this model, I am primarily interested in the effects of tree density, tree health, and biodiversity on the spread-rate of the Emerald Ash Borer. To measure these effects, this model considers the length of time it takes for the entire ash tree population in the study area to be infected and killed by the Emerald Ash Borer.  

## Model Outline
### 1) Space
In this model, the space will consist of two grids: a 2D square grid (sized according to initial parameters) without wrapping edges, housing the trees (based on their tree_id) and a corresponding array of lists housing the beetle_ids. The locations in each grid will correspond, meaning that the beetles in beetle_space[1][1] will be considered as inhabiting the tree in tree_space[1,1]. 

### 2) Actors

#### a) Beetles
Beetles are the main agent in this model which will spread outward from a randomly chosen focal point and degrade tree health to the point that the tree is killed. This model simulates their life cycle by stepping them through their life stages from egg to larvae to adult to dormant. Because reproduction is my main concern, I will only model female beetles. 
#### Properties
* _lifestage_: What stage of life the beetle is in
    * 1 = egg
    * 2 = larvae
    * 3 = dormant
    * 4 = adult
    * 5 = dead
* _num_eggs_: the number of female eggs laid by a given female. Typical range is between 50 and 90 according to literature cited in proposal. Therefore, num_eggs will be based on a randomly sampling a uniform distribution between 25 and 45 (dividing by 2 to simulate only females). 
* _egg_survival_: this represents the probability that an egg becomes a larvae.
* _winter_survival_: this represents the probability that a dormant lavae becomes an adult.
* _wanderlust_: how far an adult beetle is willing to look for a new host tree. 

For their step function, beetles actions will vary depending on life stage. First, adult beetles will find new trees to inhabit and reproduce using the _find_and_inhabit_ module. Then, population dynamics will be simulated, depending on the lifestage of the beetle, as follows:
* _Eggs_: Check _egg_survival_ to see whether egg becomes larvae or not.
* _Larvae_ will feed on their host tree (degrading its health) and then become adults. If the host tree is unable to support it, the larvae will die.
* _Dormant_ beetles will become adults. Some will not survive. 
* _Adults_ will search within their neighborhood (defined by _wanderlust_) for a new host tree. If they find one, they will lay eggs on it. Then they will die. 

#### b) Trees
Trees are the other agent in this model. They will grow and die according to their interactions with beetles.
#### Properties
* _species_: whether the tree is an ash tree (and therefore a target) or not. 
* _tree_dbh_: the size (diameter at breast height) of the tree. This will affect how many larvae it can host.
* _tree_health_: the health of the tree on a scale from 100 (perfectly healthy) to 0 (dead). 
* _has_beetles_: whether or not the tree is a host. This will only apply to ash trees, as other trees in the system will not be affected. 
* _num_eggs_: the number of female beetle eggs.
* _num_larvae_: the number of female larva.
* _num_dormant_: the number of female dormant beetles.
* _num_adult_: the number of female adult beetles.

For their step function, the following will update:
* Trees will grow based on their health. If they are currently hosting beetles, they will not grow at all, but if they are not hosts, a nominal amount will be added to _tree_dbh_. 
* _tree_health_ will decline based on the number of adults/larvae on a given tree. This decline will be small in the presence of adults (they only defoliate the tree) but larger for each larvae present. Eventually, the tree will die. 
* If a previously unaffected tree becomes infested, _has_beetles_ will be updated. 

### Institutions
At this time, this model has no institutions. 

## Initial Conditions
### Trees
* Trees will randomly be placed on a grid by sampling from a uniform discrete distribution with replacement. If we choose a cell that has already been chosen, we'll keep trying until we place all our trees.
* Trees will then be assigned a species - randomly chosen trees from our population will become ash trees. 
* Trees will have their health randomly initialized to a value from a uniform continuous distribution between .85 and 1 (the majority of trees will be healthy at the outset). 
* Trees will have their size randomly initialized based on a uniform continuous distribution. 
* All trees will, by default, have no beetles. One will be randomly chosen to be infected with one adult. 

### Beetles
* One dormant beetle will be placed on one ash tree. Then, in the first step, this beetle will become an adult who searches for a new tree to infest and subsequently, reproduces. 

## Model Parameters

Based on the above description, we need the following:
* _grid_size_: the size of our square 2D grid (i.e. square length)
* _tree_population_: the total number of trees in our system.
* _ash_prop_: the percentage of the tree population that is an ash species and therefore, susceptible. 
* _min_dbh_: the minimum dbh for trees in cm.
* _max_dbh_: the maximum dbh for trees in cm. 
* _min_health_: the minimum health of a tree (0.85)
* _max_health_: the maximum health of a tree (1) 
* _min_growth_: the minimum growth/timestep for a tree
* _max_growth_: the maximum growth/timestep for a tree
* _initial_beetles_: the number of beetles in the system at timestep 1. This defaults to 1. 
* _min_egg_survival_, _max_egg_survival_: the range of possible values of egg survival rates - varies between 0 and 1.
* _min_winter_survival_, _max_winter_survival_: the range of possible values for winter survival rates - varies between 0 and 1.

## Tree Class

Below, the class constructor initializes the tree when we call Tree(). Also, we have the following methods:
* _Tree_Growth_: checks whether a tree will grow or not based on it's infection status.
* _is_ash_: checks whether a tree is an ash or not
* _evaluate_larvae_num_: determines the maximum number of larvae a tree can support using the tree's diameter at breast height (dbh - a property of each tree) and allometric equations specific for ash trees in urban environments which estimate tree height. Using dbh and height, we can estimate surface area of phloem (the area of the tree the larvae consume) and thus, can estimate the number of larva a tree can support per square meter, based on estimations in the literature. 
* _is_infected_: determines whether tree is infected or not.
* _defoliate_decline_: randomly generated health decline due to presence of adult beetle. Different for each beetle

```python
class Tree(object):
    """
    Tree Class.
    """
    def __init__(self, model, tree_id, tree_species, tree_dbh,
                 tree_health, has_beetles, num_eggs, num_larvae,
                 num_dormant, num_adult, xloc,yloc):
        """
        constructor for tree class. By default, all trees:
        * are not ash
        * have random dbh between specified min and max (basic uniform dist)
        * have random health between specified min and max (basic uniform dist)
        * have no beetles of any lifestage
        """
        self.model = model
        self.tree_id = tree_id
        self.tree_species = tree_species
        self.tree_dbh = tree_dbh
        self.tree_health = tree_health
        self.has_beetles = has_beetles
        self.num_eggs = num_eggs
        self.num_larvae = num_larvae
        self.num_dormant = num_dormant
        self.num_adult = num_adult
        #because trees cannot move, we'll simplify later modeling by hard coding
        #tree location into their properties
        self.xloc = xloc
        self.yloc = yloc
        
        
    def tree_growth(self):
        if self.has_beetles == "Yes":
            self.tree_dbh += 0
        else:
            self.tree_dbh += numpy.random.uniform(self.min_growth,
                                                  self.max_growth)         
    
    def is_ash(self,tree_id):
        if self.model.tree_list[tree_id] == 1:
            return 1
        if self.model.tree_list[tree_id] == 0:
            return 0
            
    def evaluate_larvae_num(self, tree_id):
        """
        This method uses dbh to calculate tree height using allometric equations for urban ash trees
        (calculated by Peper et al.) and then calculates surface area of the tree. From that point, we can infer the
        number of larvae that tree can host based on the literature: 200 - 300 larva per sq. m of phloem. Therefore, this
        model will model 100-150 larvae, since we're only modeling females. 
        
        Allometric equation: height = 2.1648 + .2743(DBH)
        DBH: cm
        height: m
        Mean Square Error: 0.00002
        
        CITATIONS: 
        McCullough DG, Mercader RJ. 2012. SLAM in an urban forest: evaluation of potential strategies 
        to Slow Ash Mortality caused by emerald ash borer (Agrilus planipennis). Int. J. Pest Manag. 58:9–23.
        
        Peper, Paula J., et al. "Allometric equations for urban ash trees (Fraxinus spp.) 
        in Oakville, Southern Ontario, Canada." Urban Forestry & Urban Greening 13.1 (2014): 175-183.
        """
        #get tree dbh and transform as needed
        dbh = self.model.get_dbh(tree_id)
        dbh_m = dbh/100
        radius_m = dbh_m/2
        
        #calculate tree height
        h = 2.1648 + (.2743*dbh)
        
        #calculate surface area based on assumption that tree is right cylinder, determine
        #total larvae a tree can host, round
        available_phloem = (2*scipy.pi*(radius_m * radius_m))
        larvae_per_sqm = numpy.random.uniform(100,150)
        total_larva = round(available_phloem * larvae_per_sqm)

        return total_larva
           
        
    def is_infected(self):
        if self.num_adults > 0 or self.num_dormant > 0 or self.num_larvae >0 or self.num_eggs >0:
            return True
        else:
            return False
   

                
print "done"            
        
            
```

## Beetle Class

Below, the beetle class constructor creates a beetle every time we call Beetle(). It also houses the following methods:
* _get_location_: returns the beetle's location on the grid
* _egg_to_larva_: determines whether an egg will become a larva or whether it will die depending on a randomly chosen egg survival rate.
* _dormant_to_adult_: determines whether a dormant adult will become an egg-producing/traveling adult beetle based on a randomly chosen survival rate


```python
class Beetle(object):
    """
    Beetle Class
    """
    def __init__(self, model, lifestage, beetle_id,
                 num_eggs_laid, egg_survival, winter_survival, wanderlust):
        
        """
        constructor for beetle class. By default, there is only one
        adult beetle located on a randomly chosen ash tree. 
        """
        self.model = model
        self.lifestage = lifestage
        self.beetle_id = beetle_id
        self.num_eggs_laid = num_eggs_laid
        self.egg_survival = egg_survival
        self.winter_survival = winter_survival
        self.wanderlust = wanderlust
        
    def get_location(self):
        """
        returns beetle position, calling through model
        """
        return self.EABModel.get_beetle_location(self.beetle_id)
        
    def egg_to_larva(self):
        """
        determines whether an egg will become a larva
        """
        if numpy.random.random() <= self.egg_survival:
            return True
        else:
            return False
            
    def dormant_to_adult(self):
        """
        determines whether a dormant beetle will emerge 
        as an adult
        """
        if numpy.random.random() <= self.winter_survival:
            return True
        else:
            return False
        
    def birth_new_eggs(self, num_eggs_laid, new_host_id, x, y):
        
        for egg in range(int(num_eggs_laid)):
            
            #update tree info
            self.model.tree_list[new_host_id].num_eggs += 1
            
            #ensure beetle_id's are not duplicated
            egg_id = len(self.model.beetle_list) + egg
            
            #update beetle_space

            self.model.beetle_space[x][y].append(egg_id)
            
            
            #determine egg's life traits and add them to beetle_list
            search_rad = self.model.get_search_rad()
            self.model.beetle_list.append((self, self.beetle_id == egg_id,
                                           self.lifestage == 1, 
                                           self.num_eggs_laid == numpy.random.uniform(25, 45),
                                           self.egg_survival == numpy.random.uniform(self.model.min_egg_survival, self.model.max_egg_survival),
                                           self.winter_survival == numpy.random.uniform(self.model.min_winter_survival, self.model.max_winter_survival),
                                           self.wanderlust == search_rad))
        
    
        
print "done"

        
    
```

## Model Class

Below is the constructor (__init__ method), which initialized my model when I call EABModel(). It also has the following methods:
* _setup_space_: creates an empty grid for population
* _create_agents_: creates our trees/beetles according to model parameters 
* _infest_initial_tree_: finds a random ash tree to infest with initial beetle
* _find_neighborhood_: finds total neighborhood based on beetle position
* _find_hosts_: finds suitable trees within neighborhood
* _get_beetle_location_: find beetle location based on beetle ID
* _get_beetle_neighbors_: finds suitable trees for infestation based on beetle_ID
* _move_adults_: randomly selects from suitable trees in neighborhood, moves beetle if it's an adult
* _move_and_reproduce_: calls the _move_adults_ function, adult beetles reproduce based on beetle characteristics. 
* _grow_or_die_: simulates population dynamics of beetles.
* _step_update_: updates tree health, calls previous two functions, which call all other functions as needed. 
* _get_dbh_: gets the tree DBH
* _has_beetles_: checks whether a tree is hosting any adults, larva, or dormant beetles so we can determine tree health


```python
class EABModel(object):
    """
    Model Class.
    """
    def __init__(self,
               initial_beetles = 1,
               initial_lifestage = 3,
               tree_population = 40,
               ash_prop = 1,
               min_dbh = 13,
               max_dbh = 45,
               min_health = 75,
               max_health = 95,
               min_growth = .03, #inches per timestep
               max_growth = .07, #inches per timestep
               grid_size = 7,
               tree_health = 0.9,
               min_egg_survival = .75,
               max_egg_survival = .99,
               min_winter_survival = .7,
               max_winter_survival = .99,
               wanderlust = 5,
               ):
        
        self.beetle_list = []
       
        
        self.initial_beetles = initial_beetles
        self.initial_lifestage = initial_lifestage
        self.tree_population = tree_population
        self.ash_prop = ash_prop
        self.min_dbh = min_dbh
        self.max_dbh = max_dbh
        self.grid_size = grid_size
        self.min_health = min_health
        self.max_health = max_health
        self.min_growth = min_growth
        self.max_growth = max_growth
        self.min_egg_survival = min_egg_survival
        self.max_egg_survival = max_egg_survival
        self.min_winter_survival = min_winter_survival
        self.max_winter_survival = max_winter_survival
        self.wanderlust = wanderlust
    
        
        self.tree_list = []
        self.beetle_list = []
        
        # call setup methods to initialize model starting point.
        
        x = self.setup_space()
        print x
        y = self.create_trees()
        print y
        b = self.place_trees()
        print b
        z = self.create_beetles()
        print z
        a = self.infest_initial_tree()
        print a
    
    def get_search_rad(self):
        large_search_prob = numpy.random.random()
        if large_search_prob <= .05:
            return numpy.random.randint(round(self.grid_size*.5), self.grid_size)
        elif large_search_prob > .05:
            return numpy.random.randint(1,round(self.grid_size*.3))
        
    def setup_space(self):
        """
        creates two empty grids - one for our trees and one for our beetles. Our tree grid 
        is a simple array filled with nan, our beetle grid is an array of empty lists. 
        """
        self.tree_space = numpy.full((self.grid_size, self.grid_size),numpy.nan)
        #self.tree_space = numpy.full((self.grid_size, self.grid_size), fill_value = -1, dtype = numpy.int)
        self.beetle_space = []
        
        for row in range(self.grid_size):
            grid_row = []
            for column in range(self.grid_size):
                grid_row.append([])
                
            self.beetle_space.append(grid_row)
        return "space created"

  
    def create_trees(self):
        """
        create the agents for the model, and place them in space
        """
        # create trees with default values, specified above 
        for i in range(self.tree_population):
            self.tree_list.append(Tree(model = self, tree_id = i,
                                       tree_species = 0,
                                       tree_dbh = numpy.random.uniform(self.min_dbh, self.max_dbh),
                                       tree_health = numpy.random.uniform(self.min_health, self.max_health),
                                       has_beetles = 0,
                                       num_eggs = 0,
                                       num_larvae = 0,
                                       num_dormant = 0,
                                       num_adult = 0,
                                       xloc = 0,
                                       yloc = 0))
        return "trees created"
    
    def place_trees(self):  
        # randomly endow empty grid with trees, giving them a species
        # based on our specified ash proportion.
        tree_count = 0
        for tree in self.tree_list:
            #loop until unique
            occupied = True
            while occupied:
            #find random position
                row = numpy.random.randint(0, self.grid_size)
                col = numpy.random.randint(0, self.grid_size)
            #determine if space is empty
                if numpy.isnan(self.tree_space[row, col]):
                    occupied = False
                else:
                    occupied = True
            #determine tree species based on proportion of ash trees
            if numpy.random.random() <= self.ash_prop:
                self.tree_space[row, col] = tree.tree_id
                tree.tree_species = 1
                tree.xloc = row
                tree.yloc = col
                tree_count += 1
            else:
                self.tree_space[row, col] = tree.tree_id
                tree.tree_species = 0
                tree.xloc = row
                tree.yloc = col
                tree_count += 1
            
        return tree_count
    
    def create_beetles(self):
                
        # create beetles based on model parameters
        for beetle_id in range(self.initial_beetles):
            search_rad = self.get_search_rad()
            self.beetle_list.append(Beetle(self, beetle_id = beetle,
                                           lifestage = self.initial_lifestage, 
                                           num_eggs_laid = numpy.random.uniform(25, 45),
                                           egg_survival = numpy.random.uniform(self.min_egg_survival, self.max_egg_survival),
                                           winter_survival = numpy.random.uniform(self.min_winter_survival, self.max_winter_survival),
                                           wanderlust = search_rad))
            
        return "Initial Beetle Created"
        
    def infest_initial_tree(self):
        """
        choose an ash tree to be randomly infested with each initial beetle as specified in model
        parameters. Keeps randomly choosing trees until an ash tree is found. 
        """
        for beetle in self.beetle_list:
            ## choose random tree, check if it is an ash. If not, choose another random one 
            ashtest = 0     
            while ashtest == 0:
                random_infested = numpy.random.choice(range(len(self.tree_list)))  
                ashtest = self.tree_list[random_infested].tree_species
                if ashtest == 1:
                    self.tree_list[random_infested].num_dormant += 1
                    self.tree_list[random_infested].has_beetles = 1
                    # add beetle to grid cell corresponding to where tree is located in our beetle_space.
                    x = self.tree_list[random_infested].xloc
                    y = self.tree_list[random_infested].yloc
                    self.beetle_space[x][y].append(beetle)
                    return "initial tree is infested"
                else:
                    ashtest == 0
                                
           
    def get_dbh(self, tree_id):
        """
        returns the DBH of a tree, based on tree_id
        """
        return self.tree_list[tree_id].tree_dbh
    

    def find_neighborhood(self, beetle, x, y):
        """
        finds neighbors, accounting for grid edges, returns list of cells 
        locations within neighborhood (neighborhood type = extended Moore)
        """
        neighbor_cells = []
        search_rad = beetle.wanderlust
        x_loc = x
        y_loc = y
        
        
        try:
            for x, y in itertools.product(range(x_loc-search_rad, x_loc + search_rad + 1 ),
                                          range(y_loc-search_rad, y_loc + search_rad+ 1 )):
            
                loc_tup = (x, y)
                
                neighbor_cells.append(loc_tup)
                
        except IndexError:
            pass
        
        # remove locations outside our study area
        index = -1
        for x,y in neighbor_cells:
            index += 1
            if x < 0 or y < 0:
                neighbor_cells.remove(index)
        return neighbor_cells
    
    
    def find_hosts(self, beetle, x , y):
        """
        finds all ash trees within neighborhood, returns their locations
        as a list. Potential hosts determined based on beetle's current location 
        and their flight length ability (wanderlust)
        """
    
        neighbor_cells = self.find_neighborhood(beetle, x, y)
        potential_hosts = []
        
        for a, b in neighbor_cells:
            if a == x and b == y:
                potential_hosts.append(tree_id)
            
            if not numpy.isnan(self.tree_space[a, b]):
                
                tree_id = int(self.tree_space[x,y])
                #ashtest = tree.ash_test(tree_id)
                tree_species = self.tree_list[tree_id].tree_species
                if tree_species == 1:
                #only alive ash trees can be hosts
                    if self.tree_list[tree_id].tree_health > 0:
                        potential_hosts.append(tree_id)
                    
            return potential_hosts
        
    
    def get_beetle_location(self, beetle):
        """
        get the position of a beetle based on a query of their beetle_id 
        in self.beetle_space
        """
        row_index = -1
        
        location = []
        for row in self.beetle_space:
            row_index += 1
            col_index = -1
            for col in row:
                col_index += 1
                if beetle in col:
                    location.append(row_index)
                    location.append(col_index)
                    return location
                else:
                    continue
            
    
    def move_adults(self, beetle, x, y):
        """
        finds all ash trees within range of beetle and randomly selects one for the beetle
        to move to. If the beetle cannot find a new tree, it stays put. If the beetle moves, 
        it is removed from the old tree and added to the new one. Once settled into its new 
        home, the adult beetle will lay eggs, calling a Beetle method. 
        """
        #get beetle location
        #x, y = self.get_beetle_location(self, beetle_id)
        
        # returns list of tree_ids for potential hosts based on beetle's search radius, tree species,
        # and tree health
        x = int(x)
        y = int(y)
        potential_hosts = self.find_hosts(beetle,x, y)
        
        
        #determine how much this beetle can injure a tree by defoliating its leaves. Adult impact estimated
        # to be 1/8 or larval impact
        defoliate_stress = numpy.random.uniform(0.01, 0.025)
        
        #choose a new host from the potentials. If there is nowhere to go, the adult has a 30% chance
        #of dying while searching for a new tree to inhabit and a 70% chance of finding their way back to the 
        #original host. If they die, the beetle lifestage parameter is updated back in the step method.
        
        #if no new hosts found: beetle will either die or stay put
        if potential_hosts == None:
            stay_prob = numpy.random.uniform()
            if stay_prob <= 1:
                return "Beetle Died"  
            else: 
                #beetle doesn't move - define new location as old location
                return "beetle didn't move"
                new_host_x = x
                new_host_y = y
                new_host_id = self.get_beetles_tree(beetle)
            
                
        #if beetle finds new hosts, it randomly chooses one to infest       
        else:
            #choose randomly from potential hosts
            host_index = numpy.random.choice(range(len(potential_hosts)))
            host_index = int(host_index)
            
            #use random choice to determine tree_id
            new_host_id = potential_hosts[host_index]
            new_host_id = int(new_host_id)
            
            #use tree_id to determine beetle's new location
            new_host_x = self.tree_list[new_host_id].xloc
            new_host_y = self.tree_list[new_host_id].yloc
            return "Beetle Found New Home at " + str(new_host_x) + " " + str(new_host_y)
         
            
        #add beetle to new location
        self.beetle_space[new_host_x][new_host_y].append(beetle_id)
        self.tree_list[new_host_id].has_beetles = 1
        self.tree_list[new_host_id].num_adult += 1
            
        #remove beetle from old location
        self.beetle_space[x][y].remove(beetle_id)
            
        #add new eggs to tree
        beetle.birth_new_eggs(beetle.num_eggs_laid, new_host_id, new_host_x, new_host_y)

            
        #adult defoliates tree, leading to minimal health loss
        self.tree_list[new_host_id].tree_health -= defoliate_stress
        
        
   
    
    def get_beetles_tree(self, beetle):
        
        x, y = self.get_beetle_location(beetle)
        
        return int(self.tree_space[x, y])
        
        #for tree in self.tree_list:
            #if x == tree.xloc and y == tree.yloc:
                #return tree.tree_id
            
    def move_and_reproduce(self, beetle): 
        

        #use beetle location to find the tree they're infesting
        tree_id = self.get_beetles_tree(beetle)
            
            
        x = int(self.tree_list[tree_id].xloc)
        y = int(self.tree_list[tree_id].yloc)
            
            
        move_outcome = self.move_adults(beetle, x, y)
        if move_outcome == "Beetle Died":
            beetle.lifestage += 2
        
        print move_outcome
                                                   
                    
    def grow_or_die(self, beetle):
        """
        updates tree properties based on presence/absence of beetles. The following are updated:
        (1) if tree is uninfected, dbh increases
        (2) if tree is infected, health decreases minimally according to num_adults (defoliation) 
            and health decreases more rapidly according to num_larvae.
        (3) if tree health reaches zero, kill tree. 
        
        updates beetle lifestages:
        (1) a proportion of eggs become larva based on egg_survival. Update tree.larva_num 
        (2) a proportion of larva become dormant according to tree.evaluate_larvae_num. Update tree.dormant_num
        (3) a proportion of dormant beetles become adults. Update tree.adult_num. These new adults will find a new tree
            to colonize and then lay eggs 
        (4) if lifestage = 4 (beetle is adult), they die. 
        (5) if tree dies, kill eggs, larva, and dormant. dormant beetles that become adults will search
            for new tree to colonize. 
        """   
            
        #use beetle location to find the tree they're infesting
        tree_id = self.get_beetles_tree(beetle)
            
        #find out how many larva that tree can support
        num_larva = self.tree_list[tree_id].evaluate_larvae_num(tree_id)
            
        x = int(self.tree_list[tree_id].xloc)
        y = int(self.tree_list[tree_id].yloc)
            
        if self.tree_list[tree_id].tree_health <= 0:
            beetle.lifestage = 5  
                
        elif self.tree_list[tree_id].tree_health > 0:
                        
            ## skip dead beetles     
            
            ## ADULT BEETLES --> DEAD BEETLES 
            if beetle.lifestage == 4:
                 #kill adult beetles, update tree.num_adult, remove dead from list
                self.tree_list[tree_id].num_adult -= 1
                beetle.lifestage += 1  
                return "adult to dead"
                
            ## (EGGS --> LARVAE) or (EGGS --> DEAD)   
            elif beetle.lifestage == 1:
                self.tree_list[tree_id].num_egg -= 1
                
                #hatch eggs & kill a proportion of them due to environmental issues; update tree.num_larvae,
                #update tree_health based on # new larva, kill a proportion of larva
                survive_1 = beetle.egg_to_larva()
                #survive_1 = Beetle.egg_to_larva(self)
                if survive_1 is True:
                    beetle.lifestage += 1
                    self.tree_list[tree_id].num_larvae += 1
                    larval_decline = numpy.random.uniform(.09, .2) 
                    #range based on calibration of average tree size, average # larva/sq m, and average time
                    #from infestation that tree dies - I just played around with values until I got something reasonable.
                    # this declines the tree's health from the new larva.
                    self.tree_list[tree].tree_health -= larval_decline
                    return "egg to larva"
                else:
                    beetle.lifestage += 4 #dead
                    return "egg to dead"
                    
            ## (LARVAE --> DORMANT) or (LARVAE --> DEAD)
            elif beetle.lifestage == 2:
                self.tree_list[tree_id].num_larvae -= 1
                
                # check to see if tree can host more larva. If it can,
                # larvae goes to dormant stage. If not, it dies. Kill a proportion
                # of the larva regardless because it is inevitable that a few will die.
                if self.tree_list[tree_id].num_larvae < num_larva:
                    if numpy.random.uniform() <= .92:
                        beetle.lifestage += 1
                        self.tree_list[tree_id].num_dormant += 1
                        return "larva to dormant"
                    else:
                        beetle.lifestage += 3 #dead
                        return "larva to dead"
                    
                #tree cannot host any new larva        
                elif self.tree_list[tree_id].num_larvae > num_larva:
                    beetle.lifestage += 3
                    return "larva to dead"
        
                
            ## DORMANT --> ADULTS    
            elif beetle.lifestage == 3:
                self.tree_list[tree_id].num_dormant -= 1
                    
                if beetle.dormant_to_adult() is False:
                    beetle.lifestage += 2 #dead"
                    return "dormant to dead"
                    
                elif beetle.dormant_to_adult() is True:
                    beetle.lifestage += 1
                    return "dormant to adult"
                        

                    
                            


    def step_update(self):
        
        # for every beetle in our list:
        for beetle in self.beetle_list:
            
            #make specified beetle grow one lifestage. This will also
            #harm the tree if the beetle turns into a larva
            x = self.grow_or_die(beetle)
            print x
        #because we don't want     
        for beetle in self.beetle_list:
            #if the beetle is an adult, move it.
            if beetle.lifestage == 4:
                self.move_and_reproduce(beetle)
            else:
                continue
                
                
        

               
        
        
    
    
print "done"



```
